# CNN Feature Extraction with Filters and Pooling

```python
# importing the required libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Generate a synthetic grayscale image (since no image was uploaded)
image = np.random.randint(0, 256, (256, 256), dtype=np.uint8)
```

```python
# Define Sobel filters from the provided image
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]], dtype=np.float32)

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]], dtype=np.float32)

# Apply Sobel filter using OpenCV
sobel_x_result = cv2.filter2D(image, -1, sobel_x)
sobel_y_result = cv2.filter2D(image, -1, sobel_y)
```

```python
# Display results
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_x_result, cmap='gray')
plt.title("Sobel-X Edge Detection")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_y_result, cmap='gray')
plt.title("Sobel-Y Edge Detection")
plt.axis('off')

plt.show()
```

```python
# Task 2: Implement Max Pooling and Average Pooling using NumPy
```

```python
# Create a random 4x4 matrix
np.random.seed(42)  # For reproducibility
input_matrix = np.random.randint(0, 10, (4, 4))
```

```python
# Function for max pooling
def max_pooling(matrix, pool_size=2):
    pooled_matrix = np.zeros((matrix.shape[0] // pool_size, matrix.shape[1] // pool_size))
    for i in range(0, matrix.shape[0], pool_size):
        for j in range(0, matrix.shape[1], pool_size):
            pooled_matrix[i // pool_size, j // pool_size] = np.max(matrix[i:i + pool_size, j:j + pool_size])
    return pooled_matrix
```

```python
# Function for average pooling
def average_pooling(matrix, pool_size=2):
    pooled_matrix = np.zeros((matrix.shape[0] // pool_size, matrix.shape[1] // pool_size))
    for i in range(0, matrix.shape[0], pool_size):
        for j in range(0, matrix.shape[1], pool_size):
            pooled_matrix[i // pool_size, j // pool_size] = np.mean(matrix[i:i + pool_size, j:j + pool_size])
    return pooled_matrix
```

```python
# Apply pooling operations
max_pooled_matrix = max_pooling(input_matrix)
avg_pooled_matrix = average_pooling(input_matrix)
```

```python
# Print the results
print("Original 4x4 Matrix:\n", input_matrix, "\n")
print("Max-Pooled 2x2 Matrix:\n", max_pooled_matrix, "\n")
print("Average-Pooled 2x2 Matrix:\n", avg_pooled_matrix, "\n")
```
